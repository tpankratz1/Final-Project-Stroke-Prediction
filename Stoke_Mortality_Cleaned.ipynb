```python
import pandas as pd
import numpy as np
```

```python
# Import and read the Stroke Mortality dataset
stroke_df = pd.read_csv("Data/Stroke_Mortality_Data.csv")
stroke_df.head()
```

```python
# health_rankings_final_df["State"].replace("Alabama", "AL", inplace=True)
```

```python
# Drop the non-beneficial columns
stroke_df = stroke_df.drop(["Year","DataSource","Class","Topic","Data_Value_Footnote_Symbol","Data_Value_Footnote","StratificationCategory1","StratificationCategory2","TopicID","Data_Value_Type","Data_Value_Unit","GeographicLevel","Y_lat","X_lon"], 1)
stroke_df.head()
```

```python
# Rename all columns for clarity
stroke_df.columns =['State', 'Description_Detail', 'Num_Per_100k', 'Gender', 'Ethnicity', 'FIPS']
stroke_df
```

```python
# Filtering the rows where Overall exists in both Gender and Ethnicity columns
stroke_df = stroke_df[stroke_df['Gender'].str.contains('Overall')]
stroke_df = stroke_df[stroke_df['Ethnicity'].str.contains('Overall')]
stroke_df
```

```python
# Check number of unique values in each column.
stroke_df.nunique()
```

```python
stroke_df = stroke_df.drop(["Gender","Ethnicity"], 1)
stroke_df.head()
```

```python
# Check for null values
stroke_df.isnull().sum()
```

```python
stroke_df["State"].unique()
```

```python
stroke_df = stroke_df[stroke_df.State != 'AS']
stroke_df = stroke_df[stroke_df.State != 'GU']
stroke_df = stroke_df[stroke_df.State != 'MP']
stroke_df = stroke_df[stroke_df.State != 'PR']
stroke_df = stroke_df[stroke_df.State != 'US']
stroke_df = stroke_df[stroke_df.State != 'VI']
stroke_df.nunique()
```

```python
stroke_df["State"].unique()
```

```python
# Move FIPS column to the far left
stroke_df = stroke_df[["FIPS","State","Description_Detail","Num_Per_100k"]]
stroke_df
```

```python
stroke_df = stroke_df.drop(["State","Description_Detail"], 1)
stroke_df
```

```python
stroke_df = stroke_df[stroke_df.FIPS >= 1000]
stroke_df
```

```python
stroke_df.set_index('FIPS')
stroke_df
```

```python
stroke_df.to_csv('Data/Stroke_Mortality_Final.csv', index=False)
```
