```python
import pandas as pd
import numpy as np
```

```python
# Import and read the Health Rankings data
health_rankings_df = pd.read_csv("Data/2018_County_Health_Rankings_US_Data.csv")
health_rankings_df.head()
```

```python
# Replace % with Percent in column headings
health_rankings_df = health_rankings_df.rename(columns=lambda s: s. replace("%", "Percent"))
health_rankings_df.head()
```

```python
# Include wanted columns and remove unwanted columns
health_rankings_df = health_rankings_df[["FIPS","State","County","Percent Smokers","Percent Obese","Food Environment Index","Percent With Access","PCP Rate","MHP Rate","Percent Some College","Percent Unemployed","Income Ratio","Violent Crime Rate","Average Daily PM2.5","Percent Long Commute - Drives Alone"]]
health_rankings_df.head()
```

```python
# Rename most column headings
health_rankings_df = health_rankings_df.rename(columns=lambda s: s.replace("Percent Smokers", "Percent_Smokers"))
health_rankings_df = health_rankings_df.rename(columns=lambda s: s.replace("Percent Obese", "Percent_Obese"))
health_rankings_df = health_rankings_df.rename(columns=lambda s: s.replace("Food Environment Index", "Food_Environment_Index"))
health_rankings_df = health_rankings_df.rename(columns=lambda s: s.replace("Percent With Access", "Percent_Access_Exercise"))
health_rankings_df = health_rankings_df.rename(columns=lambda s: s.replace("PCP Rate", "Primary_Care_Physician_Rate"))
health_rankings_df = health_rankings_df.rename(columns=lambda s: s.replace("MHP Rate", "Mental_Health_Providers_Rate"))
health_rankings_df = health_rankings_df.rename(columns=lambda s: s.replace("Percent Some College", "Percent_Some_College"))
health_rankings_df = health_rankings_df.rename(columns=lambda s: s.replace("Percent Unemployed", "Percent_Unemployed"))
health_rankings_df = health_rankings_df.rename(columns=lambda s: s.replace("Income Ratio", "Income_Inequality_Ratio"))
health_rankings_df = health_rankings_df.rename(columns=lambda s: s.replace("Violent Crime Rate", "Violent_Crime_Rate"))
health_rankings_df = health_rankings_df.rename(columns=lambda s: s.replace("Average Daily PM2.5", "Average_Daily_Air_Polution_Particle_Matter"))
health_rankings_df = health_rankings_df.rename(columns=lambda s: s.replace("Percent Long Commute - Drives Alone", "Percent_Long_Commute_Drives_Alone"))
health_rankings_df
```

```python
# Import and read the additional Health Rankings data (Percent Rural)
health_rankings_additional_df = pd.read_csv("Data/2018_County_Health_Rankings_US_Additional_Data.csv")
health_rankings_additional_df.head()
```

```python
# Replace % with Percent in column headings
health_rankings_additional_df = health_rankings_additional_df.rename(columns=lambda s: s. replace("% Rural", "Percent_Rural"))
health_rankings_additional_df.head()
```

```python
# Remove unwanted columns
health_rankings_additional_df = health_rankings_additional_df[["FIPS","Percent_Rural"]]
health_rankings_additional_df.head()
```

```python
health_rankings_additional_df.dtypes
```

```python
health_rankings_additional_df.set_index('FIPS')
health_rankings_additional_df
```

```python
health_rankings_additional_df.to_csv('Data/Health_Rankings_Additional_Final.csv', index=False)
```

```python
# Combine the the two data sets into a single dataset
health_rankings_final_df = pd.merge(health_rankings_additional_df, health_rankings_df, on=["FIPS", "FIPS"])
health_rankings_final_df.head()
```

```python
# Move Percent Rural column to the far right
health_rankings_final_df = health_rankings_final_df[["FIPS","State","County","Percent_Smokers","Percent_Obese","Food_Environment_Index","Percent_Access_Exercise","Primary_Care_Physician_Rate","Mental_Health_Providers_Rate","Percent_Some_College","Percent_Unemployed","Income_Inequality_Ratio","Violent_Crime_Rate","Average_Daily_Air_Polution_Particle_Matter","Percent_Long_Commute_Drives_Alone","Percent_Rural"]]
health_rankings_final_df
```

```python
# Check number of unique values in each column.
health_rankings_final_df.nunique()
```

```python
health_rankings_final_df.dtypes
```

```python
health_rankings_final_df.set_index('FIPS')
health_rankings_final_df
```
